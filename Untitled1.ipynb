```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from multiprocessing import Pool
import time


def f(xx):
    r=0
    for i in range(xx):
        r=r+i
    return r

n=10000
y1=[]
y2=[]
x=[x for x in range(100,2100, 100)]
for n  in  x:
    t1 =time.time()
    with Pool() as p:
        p.map(f, [x for x in range(n)])

    t2 =time.time()
    y1.append(t2-t1)
    t1 =time.time()
    for xx in range(n):
        f(xx)
    t2 =time.time()
    y2.append(t2-t1)

%matplotlib inline

plt.plot(x, y1,  color='#FB8500', label='Multi')
plt.plot(x, y2,  color='#023047', label='Single')
plt.legend()
plt.title('Оценка производительности: чем выше - тем хуже')
plt.show()
```
